## 3. Feature Engineering
We enhance the dataset by calculating the **Yield** ($Yield = \frac{Production}{Area}$) and grouping individual crops into broader categories like Cereals, Pulses, and Fruits for better analysis.

```python
# Calculate Yield
crop.loc[:, 'Yield'] = crop['Production'] / crop['Area']
```

```python
# yield per crop type
top_yield_crops = crop.groupby('Crop')['Yield'].mean().sort_values(ascending=False)
print(top_yield_crops.head(10))
```

```python
def get_crop_category(crop):
    crop = crop.lower()
    
    # Cereals 
    if any(x in crop for x in ['rice', 'paddy', 'wheat', 'maize', 'bajra', 'jowar', 'ragi', 'barley', 'millets', 'cereals', 'korra', 'samai', 'varagu']):
        return 'Cereals'
    
    # Pulses
    elif any(x in crop for x in ['pulses', 'gram', 'urad', 'moong', 'masoor', 'arhar', 'tur', 'lentil', 'khesari', 'cowpea', 'moth', 'peas & beans']):
        return 'Pulses'
    
    # Fruits
    elif any(x in crop for x in ['apple', 'banana', 'grapes', 'mango', 'orange', 'papaya', 'pineapple', 'citrus', 'fruit', 'sapota', 'litchi', 'peach', 'pear', 'plums', 'pome']):
        return 'Fruits'
    
    # Vegetables
    elif any(x in crop for x in ['vegetable', 'onion', 'potato', 'tomato', 'brinjal', 'cabbage', 'cauliflower', 'carrot', 'beet root', 'bhindi', 'gourd', 'guard', 'cucumber', 'drum stick', 'radish', 'turnip', 'yam', 'sweet potato', 'tapioca']):
        return 'Vegetables'
    
    # Oilseeds
    elif any(x in crop for x in ['oilseeds', 'groundnut', 'sesamum', 'sunflower', 'soyabean', 'castor', 'mustard', 'linseed', 'safflower', 'niger', 'coconut']):
        return 'Oilseeds'
    
    # Commercial / Plantation / Spices
    elif any(x in crop for x in ['sugarcane', 'tea', 'coffee', 'rubber', 'tobacco', 'cotton', 'kapas', 'jute', 'mesta', 'pepper', 'cardamom', 'ginger', 'garlic', 'turmeric', 'chillies', 'coriander', 'arecanut', 'cashewnut']):
        return 'Commercial'
    
    else:
        return 'Other'


# Apply categorization
crop['Crop_Category'] = crop['Crop'].apply(get_crop_category)
```

```python
# Check if the new column exists and see the counts
print(crop['Crop_Category'].value_counts())
```

## 4. Production Trends and Insights

```python
# 1. Correlation Matrix
crop[['Area', 'Production']].corr().round(2)
```

```python
# 2. Which states produce the most volume overall?
top_states = crop.groupby('State_Name')['Production'].sum().nlargest(5)

print("Top 5 States by Total Production:\n")
print(top_states.apply(lambda x: f"{x:,.0f}"))
```

```python
top_states
```

```python
# 3. Which crops take up the most physical land area?
top_area_crops = crop.groupby('Crop')['Area'].sum().nlargest(5)

print("Top 5 Crops by Total Land Area:\n")
print(top_area_crops.apply(lambda x: f"{x:,.0f}"))
```

```python
# 4. How has total production changed over the years?
yearly_production = crop.groupby('Crop_Year')['Production'].sum()

print("Total Production by Year:\n")
print(yearly_production.apply(lambda x: f"{x:,.0f}"))   
```
